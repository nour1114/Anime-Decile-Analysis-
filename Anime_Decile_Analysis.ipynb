```python
def homework(anime_data_extracted):
    import pandas as pd

    df = anime_data_extracted.copy()

    # Define decile labels
    labels = ['01','02','03','04','05','06','07','08','09','10']

    # Create deciles based on Members (01 = lowest, 10 = highest)
    df['decile'] = pd.qcut(df['Members'], q=10, labels=labels)

    # Group and aggregate
    result = df.groupby('decile').agg(
        MeanScore=('Score', 'mean'),
        TotalMembers=('Members', 'sum')
    )

    # Ensure correct order
    result = result.reindex(labels)

    # Compute cumulative percentage
    total_members = result['TotalMembers'].sum()
    result['CumulativePct'] = result['TotalMembers'].cumsum() / total_members * 100

    return result
```
